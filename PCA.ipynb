```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_rows',150)
pd.set_option('display.max_columns',150)
#下の二行でパンダスの表示される部分が150×150となった
```

```python
df= pd.read_csv('winning eleven.csv',header=0)
#csvファイルの読み込み
```

```python
df#データの確認
```

```python
x= pd.read_csv('winning eleven.csv',header=0,usecols=range(3,27))
corrmat=x.corr()
plt.figure(figsize=(30,20))
sns.heatmap(corrmat, cmap=sns.color_palette('coolwarm',10),annot=True,vmin=-1.0)
#相関行列を算出，annot=True より相関の値もわかる
#相関の値が低いときはうすくなるように設定した．この色好き
```

```python
df.iloc[:,3:27].describe().transpose()
```

```python
x_stats=x.describe().transpose()
def norm(x):
    return(x-x_stats['mean'])/x_stats['std']
x=norm(x)
#データの正規化をして主成分分析を行なっている
```

```python
x
#正規化されたxの確認
```

```python
import sklearn
from sklearn.decomposition import PCA
pca=PCA()
pca.fit(x)
feature=pca.transform(x)
#主成分分析の実行
```

```python
np.dot(feature,pca.components_)
#これより，feature(要約データ)×pca.components_(回転行列)=x(もともとのデータ)がわかる
```

```python
pd.DataFrame(feature, columns=["PC{}".format(x+1) for x in range(len(x.columns))]).head()
#要約データの値，なぜか4行までしか表示されていないが128列存在している
```

```python
annotations=df.iloc[:,1]
plt.figure(figsize=(10,10))
X=feature[0:127,0]
Y=feature[0:127,1]
plt.scatter(X[0:63],Y[0:63],c='red',label='CMF')
plt.scatter(X[64:127],Y[64:127],c='blue',label='DMF')
plt.legend()
plt.xlabel("PCA1")
plt.ylabel("PCA2")
for i, label in enumerate(annotations):
    plt.annotate(label, (X[i], Y[i]))
#ようやくされたデータを可視化,名前を表示するためにannotationを使用
```

```python
pd.DataFrame(pca.explained_variance_ratio_, index=["PC{}".format(x + 1) for x in range(len(x.columns))], columns =['contribution ratio'])
#相関行列の固有値を算出
```

```python
ratio=np.cumsum(pca.explained_variance_ratio_)
plt.plot(ratio,"-o")
plt.xlabel("principal componet")
plt.ylabel("cumulative contribution ratio")
#固有値を大きい順番に累積分布にさせた,7こ目でようやく0.8を超えていることがわかる.
```

```python
# PCA の固有値
pd.DataFrame(pca.explained_variance_, index=["PC{}".format(x + 1) for x in range(len(x.columns))],columns=['eigenvalue'])
```

```python
# PCA の固有ベクトル
a=pd.DataFrame(pca.components_,columns=[x.columns],index=["PC{}".format(x + 1) for x in range(len(x.columns))])
a.iloc[0:5,:]
#回転行列の確認,5行目まで表示
```

```python
y=a.iloc[0:5,:].transpose()
y
```

```python
for i in range(24):
    for j in range(5):
        y.iloc[i,j]=y.iloc[i,j]/(y.iloc[i,0]**2+y.iloc[i,1]**2)**0.5 
#回転行列をpromax回転（斜こう回転）①
```

```python
#回転行列をpromax回転（斜こう回転）②
y=y/np.amax(np.abs(y))
```

```python
#回転行列をpromax回転（斜こう回転）③
for i in range(24):
    for j in range(5):
        y.iloc[i,j]=y.iloc[i,j]**3
y=y.transpose()
pd.DataFrame(y)
```

```python
#feature(要約データ)×pca.components_(回転行列)=x(もともとのデータ)→F×P=X
#F=X×P転置(P×P転置)-1とすればpromax回転後の要約データが手に入る
z=np.dot(y.transpose(),np.linalg.inv(np.dot(y,y.transpose())))
feature_rot=np.dot(x,z)
feature_rot=pd.DataFrame(feature_rot)
feature_rot
```

```python
annotations=df.iloc[:,1]
plt.figure(figsize=(10,10))
X_rot=feature_rot.iloc[0:127,0]
Y_rot=feature_rot.iloc[0:127,1]
plt.scatter(X_rot[0:63],Y_rot[0:63],c='red',label='CMF')
plt.scatter(X_rot[64:127],Y_rot[64:127],c='blue',label='DMF')
plt.legend()
plt.xlabel("PCA1")
plt.ylabel("PCA2")
for i, label in enumerate(annotations):
    plt.annotate(label, (X_rot[i], Y_rot[i]))
#promax回転後の要約されたデータを可視化，正直あまり変わらんかも
```
